```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
data = pd.read_csv('advertising.csv')
print(data.head())
```

```python
x = data[['TV']]
y = data['Sales']
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
# Value of sales prediction using TV 

a = model.intercept_
print(a)

b = model.coef_
print(b)
```

# Multiple Linear Reggression

```python
# Importing libraries

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt

import pandas as pd

# Importing the data file
df = pd.read_csv("students_mark.csv")
df.head()

# Setting the features as x and y with their respective values
x = df[['Hours', 'Assignments']]

y = df['Marks']

print("The value of x is:", x)

print("The value of y is:", y)

#Train split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state = 5)

# Fitting in the model
model = LinearRegression()
model.fit(x_train, y_train)

#Predicting the outcomes
ypredict = model.predict(x_test)
print("The further predicted values for  y are:", ypredict)

#Evaluating the model
mse = mean_squared_error(y_test,ypredict)
print(f"MSE Value is{mse:.2f}")

mae = mean_absolute_error(y_test, ypredict)
print(f"MAE Value is{mae:.2f}")

```
