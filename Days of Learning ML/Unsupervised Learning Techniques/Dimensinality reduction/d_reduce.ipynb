# Dimensinality Reduction Begins From Here

```python
# Importing the libraries

import pandas as pd

# Reading the data

data = pd.read_csv("wine.data.csv")

data.head()
```

```python
# In Here we convert the higher no. of columns into fewer using PCA

from sklearn.decomposition import PCA

from sklearn.preprocessing import StandardScaler

```

```python
# Scaling the values

scaler = StandardScaler()
df_scaled = scaler.fit_transform(data)

df_scaled

df_scal = pd.DataFrame(df_scaled ,columns= range(1,15))
df_scal
```

```python
# CHecking the values of class

data.Class.unique()
```

```python
# Dropping off the Class 

df_scal.drop(columns = [1])
```

```python
df_scal.head()
```

```python
# Using the library

principal = PCA()
component = principal.fit_transform(df_scal)
component
```

```python
component = pd.DataFrame(component, columns = range(1,15))
component.head()
```

```python
# Checkinng the eigen values

eigen_values = principal.explained_variance_
print(eigen_values)
```

```python
# Checking the percentage for each columns

p_value = principal.explained_variance_ratio_
print(p_value)
```
