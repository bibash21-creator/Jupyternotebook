# Example Case for Diabetes Data Set

```python
# Importing the dataset

dfile = pd.read_csv("Diabetes.csv")
dfile.head()
```

```python
# SPlitting the features

x = dfile[['Glucose', "Age"]]
y = dfile.Diabetes
```
