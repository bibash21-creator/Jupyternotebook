# Decision Tree Starts Here

```python
# Importing the libraries

from sklearn.tree import DecisionTreeClassifier

import pandas as pd

from sklearn.metrics import accuracy_score, confusion_matrix

from sklearn.model_selection import train_test_split

```

```python
# Importing the data

dfile = pd.read_csv("Iris.csv")
dfile.head()
```

```python
# Separating the features

x = dfile.drop(['Species', 'Id'], axis=1)
y = dfile.Species
```

```python
x.head()
```

```python
y.head()
```

```python
# Training the model

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
# Fitting the model

dtree_model = DecisionTreeClassifier()
dtree_model.fit(x_train, y_train)
```

```python
# Predicting the outcomes

ypredict= dtree_model.predict([[5,4,1,2]])
ypredict
```

```python
# Depicitng the values in decision trees

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

```

```python
# Features_name is name of columns

plt.figure(figsize = (15,20))

plot_tree(dtree_model, feature_names = x.columns, class_names= y.unique(), filled=True)
plt.show()
```
