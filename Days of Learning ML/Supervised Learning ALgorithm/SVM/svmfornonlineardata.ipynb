# SVM for Non - Linear Data Set

```python
# Importing the libraries

import pandas as pd

from sklearn.metrics import accuracy_score, confusion_matrix

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler

from sklearn import svm


import seaborn as sbn

import matplotlib.pyplot as plt
```

```python
# Imprting the data set

dfile = pd.read_csv("linear_unseparable.csv")
dfile.head()
```

```python
# Descrbing the dfile and separating the features

dfile.describe()

x = dfile[['F1', 'F2']]

y = dfile['class']
```

```python
# Training the data

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 5)
```

```python
# Scaling the values

scaler = StandardScaler()

xtrain_s = scaler.fit_transform(x_train)
```

```python
# Fitting the model

svm_model = svm.SVC(kernel='rbf')
svm_model.fit(xtrain_s, y_train)
```
